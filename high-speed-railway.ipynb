```python
# Setup
!pip install pandas numpy
import pandas as pd

df0 = pd.read_csv('railway_delays_2020_01_1.csv')
df0.head()
weather = pd.read_csv('weather.csv')
```

```python
#Uses a list comprehension to create a list of the filenames 
# from January 1st, 2020  to January 27th,2020

filenames = ['railway_delays_2020_01_' + str(i) + '.csv' for i in range(1,28)]
print(filenames)
```

```python
#Use a list comprehension to create a list of DataFrames named railway_delays_dfs

railway_delays_dfs = [pd.read_csv(filename) for filename in filenames]

#concatenates multiple datasets of railway_delays_2020 and prints summary information about it

railway_delays_full = pd.concat(railway_delays_dfs)

railway_delays_full.info()

```

```python
for df in railway_delays_dfs: 
    df.columns = df.columns.str.lower() 
    df.columns = df.columns.str.strip() 


railway_delays_full = pd.concat(railway_delays_dfs) 
railway_delays_full.info()
```

```python
for col in railway_delays_full.columns:

    #uppercase and strip any hidden extra whitespace in all the object datatype columns

    if railway_delays_full[col].dtypes == 'object': 
        railway_delays_full[col] = railway_delays_full[col].str.upper()
        railway_delays_full[col] = railway_delays_full[col].str.strip() 

    #convert all float datatype columns to integers
    elif railway_delays_full[col].dtypes == 'float64':
        railway_delays_full[col] = railway_delays_full[col].astype('int64')


railway_delays_full.info()
```

```python
#Impact of Holiday Travel
#amount of train rides occurred on major holidays 

railway_delays_full_mayor_holiday_count_true = railway_delays_full['major_holiday'].value_counts()[True]

railway_delays_full_mayor_holiday_count_false = railway_delays_full['major_holiday'].value_counts()[False]
#print('The amount of train rides occurred on major holidays is: ', railway_delays_full_mayor_holiday_count_true)
#print('The amount of train rides occurred on non-holidays is: ', railway_delays_full_mayor_holiday_count_false)

#calculates the average of arrival_delay and departure_delay
railway_delays_full_by_major_holiday = railway_delays_full.groupby(['major_holiday']).agg({'arrival_delay':'mean','departure_delay':'mean',}).reset_index()
column_mapper = {'arrival_delay' : 'avg_arrival_delay', 'departure_delay':'avg_departure_delay' }
railway_delays_full_by_major_holiday = railway_delays_full_by_major_holiday.rename(mapper=column_mapper, axis=1)

railway_delays_full_by_major_holiday
```

```python
#Compute descriptive summary statistics on the distances from station to station
railway_delays_full['mileage'].describe()

```

```python
#checks if longer-than-average distances correspond to longer delays

railway_delays_full['long_distance'] = railway_delays_full['mileage'] > 88
railway_delays_full.head()
```

```python
#calculate the average arrival_delay and departure_delay

railway_delays_arrival_departure_avg_delay = railway_delays_full.groupby(['long_distance']).agg({'arrival_delay':'mean','departure_delay':'mean',}).reset_index()

column_mapper = {'arrival_delay' : 'avg_arrival_delay', 'departure_delay':'avg_departure_delay' }
railway_delays_arrival_departure_avg_delay = railway_delays_arrival_departure_avg_delay.rename(mapper=column_mapper, axis=1)

railway_delays_arrival_departure_avg_delay
```

```python
#clean weather data


for col in weather.columns:

    #uppercase and strip any hidden extra whitespace in all the object datatype columns

    if weather[col].dtypes == 'object': 
        weather[col] = weather[col].str.upper()
        weather[col] = weather[col].str.strip() 

    #convert all float datatype columns to integers
    elif weather[col].dtypes == 'float64':
        weather[col] = weather[col].astype('int64')


railway_delays_full.info()

```

```python
#Merge weather with railway_delays_full using the ride_id column

railway_delays_join_weather = pd.merge(left = weather,
right = railway_delays_full,
left_on = 
'ride_id',
right_on = 
'ride_id',
how = 'inner')

railway_delays_join_weather.head()

```

```python
#calculate the average of arrival_delay and departure_delay because of weather.
#Sort the result from highest arrival_delay to lowest.

railway_delays_join_weather_arr_dep_avg_delay = railway_delays_join_weather.groupby(['weather']).agg({'arrival_delay':'mean','departure_delay':'mean',}).reset_index()
 
column_mapper = {'arrival_delay' : 'avg_arrival_delay', 'departure_delay':'avg_departure_delay' }
railway_delays_join_weather_arr_dep_avg_delay = railway_delays_join_weather_arr_dep_avg_delay.rename(mapper=column_mapper, axis=1)

railway_delays_join_weather_arr_dep_avg_delay.sort_values(by='avg_arrival_delay', ascending=False)
```
